```python
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
  df1[col].replace({'Yes': 1, 'No': 0}, inplace=True) # Replaces Yes & No with (1, 0)
```

```python
for col in df1:
  print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female': 1, 'Male': 0}, inplace=True)
```

```python
df1['gender'].unique()
```

```python
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns # one hot encoding
```

```python
df2.sample(4)
```

```python
df2.dtypes
```

**2. SCALING**

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges'] # As these cols are not in the format of 0 or 1, that's why we're scaling it so that they come in 0 and 1 range

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
  print(f'{col} : {df2[col].unique()}')
```

**3. TRAIN AND TEST SPLIT**

```python
X = df2.drop('Churn', axis='columns') # All the columns except Churn
y = df2['Churn'] # is Churn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

**4. Build a model (ANN) in tensorflow/keras**

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(25, input_shape=(26,), activation='relu'), # No. of neurons for input layer and hidden layer
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid') # is sigmoid cause output is 1 & 0
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred = []
for element in yp:
  if element > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:5]
```

**5. CLASSIFICATION REPORT AND CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
